```python
import pymongo
from furl import furl
from collections import Counter

client = pymongo.MongoClient('mongodb://mongodb:itstimeformongodb@68.183.76.109:27017/')
db = client.crawlerSearchLinks
collection = db.searchItems

domain_counter = Counter()
for document in client.crawlerSearchLinks.searchItems.find({}):
    url = furl(document['URL'])
    domain_counter.update([url.origin])

len(domain_counter)
```

```python
from frozendict import frozendict

grouped_by_inn = {}
for document in client.crawlerSearchLinks.searchItems.aggregate([
    {
        '$group': {
            '_id': "$INN",
            'urls': {'$push': '$URL'}
        }
    }
]):
    grouped_by_inn[document['_id']] = document['urls']

```

```python
i = 0
filtered_by_inn = {}
for inn, urls in grouped_by_inn.items():
    filtered_by_inn[inn] = list(
        filter(
            lambda url: domain_counter[furl(url).origin] <= 10,
            urls
        )
    )
    if not filtered_by_inn[inn]:
        del filtered_by_inn[inn]

```
